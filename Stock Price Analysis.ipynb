```python
import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
start = datetime.datetime(2013,1,1)
end = datetime.datetime(2016,1,1)
```

```python
tesla = web.DataReader("TSLA",'yahoo',start,end)
gm = web.DataReader("GM", 'yahoo', start, end)
ford = web.DataReader("F", 'yahoo', start, end)
```

```python
tesla.to_csv('Tesla_stock.csv')
gm.to_csv('GM_stock.csv')
ford.to_csv('Ford_stock.csv')
```

# Stock Price Visualization

```python
tesla['High'].plot(Label = 'TESLA High price',figsize=(15,7))
tesla['Low'].plot(Label = 'TESLA Low price')
tesla['Open'].plot(Label = 'TESLA Open price')
tesla['Close'].plot(Label = 'TESLA Close price')
plt.xlabel('Dates')
plt.ylabel('Stock price')
plt.title('Tesla stock price')
plt.legend()
plt.show()
```

```python
tesla['Open'].plot(label='Tesla', figsize=(15,7))
ford['Open'].plot(label='Ford')
gm['Open'].plot(label='Gm')
plt.title('Opening stock prices')
plt.ylabel('Prices')
plt.legend()
plt.show()
```

# Total Money Traded = Open prices * Volume of stocks traded

```python
# Adding a column of TOtal money traded
tesla['Total Money Traded'] = tesla['Open']*tesla['Volume']
ford['Total Money Traded'] = ford['Open']*ford['Volume']
gm['Total Money Traded'] = gm['Open']*gm['Volume']
```

```python
ford['Open'].plot(figsize = (15,7))
ford['MA10']= ford['Open'].rolling(10).mean()
ford['MA10'].plot(label = 'MA10')
ford['MA50']= ford['Open'].rolling(50).mean()
ford['MA50'].plot(label = 'MA50')
ford['MA100']= ford['Open'].rolling(100).mean()
ford['MA100'].plot(label = 'MA100')
plt.legend()
```

```python
#Scatter plot to compare Opening stock prices of all three automobile companies
from pandas.plotting import scatter_matrix
```

```python
automobile_comp = pd.concat([tesla['Open'],gm['Open'],ford['Open']],axis =1)
automobile_comp.columns=('Tesla','GM', 'Ford')
```

```python
scatter_matrix(automobile_comp,figsize=(10,10),hist_kwds = {'bins':20})
```

# Daily Percentage Change

```python


tesla['Returns'] = (tesla['Close']/tesla['Close'].shift(1))-1
gm['Returns'] = (gm['Close']/gm['Close'].shift(1))-1
ford['Returns'] = (ford['Close']/ford['Close'].shift(1))-1
```

```python
tesla['Returns'].hist(bins = 100,label ='Tesla',alpha =0.5,figsize=(15,7))
ford['Returns'].hist(bins = 100,label='Ford',alpha = 0.5)
gm['Returns'].hist(bins = 100, label = 'GM', alpha= 0.5)
plt.legend()
```

```python
tesla['Returns'].plot(kind = 'kde',label = 'Tesla',figsize=(15,7))
ford['Returns'].plot(kind = 'kde',label = 'Ford')
gm['Returns'].plot(kind = 'kde',label = 'GM')
plt.legend()
```

```python
returns_data = pd.concat([tesla['Returns'], gm['Returns'], ford['Returns']],axis = 1)
returns_data.columns=['Tesla Returns', 'Ford Returns', 'GM Returns']
returns_data.plot(kind= 'box', figsize =(15,7))
```
